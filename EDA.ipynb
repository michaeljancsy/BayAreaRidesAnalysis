```python
import numpy as np
np.random.seed = 0
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# load pandas dataframe
df = pd.read_pickle('dataframe.pkl')
```

# Rating Clustering

```python
from sklearn.cluster import KMeans
from sklearn.cluster.hierarchical import AgglomerativeClustering
from sklearn.metrics import silhouette_score
```

```python
rating_columns = df.columns[df.columns.str.startswith('rating')]
X_rating = df[rating_columns]
```

```python
# K Means
k_vals = range(2, 10)
silhouette_scores = []
inertias = []

for k in k_vals:
    test_kmeans = KMeans(n_clusters=k, n_init=50)
    labels = test_kmeans.fit_predict(X_rating)
    silhouette_scores.append(silhouette_score(X_rating, labels))
    inertias.append(test_kmeans.inertia_)
    
plt.plot(k_vals, inertias)
```

```python
plt.plot(k_vals, silhouette_scores)
```

```python
# train final k means model
final_kmeans = KMeans(n_clusters=3, n_init=50)
kmeans_labels = final_kmeans.fit_predict(X_rating)
```

```python
# look at cluster centers
df['kmeans_label'] = kmeans_labels
kmeans_cluster_centers = pd.DataFrame(final_kmeans.cluster_centers_, columns=rating_columns)
kmeans_cluster_centers
```

```python
sns.heatmap(kmeans_cluster_centers)
```

```python
df.groupby('kmeans_label').count()[['name']]
```

Cluster 1 - fun, aerobic, scenic, not technical

# NLP

```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.decomposition import NMF, LatentDirichletAllocation


def print_top_words(model, feature_names, n_top_words):
    """Function copied from sklearn docs"""
    for topic_idx, topic in enumerate(model.components_):
        message = "Topic #%d: " % topic_idx
        message += " ".join([feature_names[i]
                             for i in topic.argsort()[:-n_top_words - 1:-1]])
        print(message)
    print()
```

```python
# featurize with TF-IDF
tfidf_vectorizer = TfidfVectorizer(stop_words='english', ngram_range=(1,3), max_features=250)
X_tfidf = tfidf_vectorizer.fit_transform(df['desc_long'])
```

```python
# decompose with NMF
errors = []
n_topics_range = range(2, 40)
for n_topics in n_topics_range:
    test_nmf = NMF(n_components=n_topics, alpha=0, init='nndsvd')
    test_nmf.fit(X_tfidf)
    errors.append(test_nmf.reconstruction_err_)
```

```python
plt.plot(n_topics_range, errors)
```

```python
nmf = NMF(n_components=10, alpha=0, init='nndsvd')
X_nmf = nmf.fit_transform(X_tfidf)
```

```python
print_top_words(nmf, tfidf_vectorizer.get_feature_names(), 10)
```

```python
# LSA (truncated SVD)
from sklearn.decomposition import TruncatedSVD
```

```python
svd = TruncatedSVD(n_components=10)
X_svd = svd.fit_transform(X_tfidf)
```

```python
print_top_words(svd, tfidf_vectorizer.get_feature_names(), 10)
```

```python
# compare NMF results to quantitative features
```

```python
# add topic features to dataframe
n_topics = nmf.components_.shape[0]
nmf_columns = ['nmf_topic'+str(i) for i in range(n_topics)]
df_nmf = pd.DataFrame(X_nmf, columns=nmf_columns)
svd_columns = ['svd_topic'+str(i) for i in range(n_topics)]
df_svd = pd.DataFrame(X_svd, columns=svd_columns)
df = df.join(df_nmf)
df = df.join(df_svd)
```
